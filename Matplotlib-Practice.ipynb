```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=list(range(7))
y=[50,51,52,48,47,49,46]
```

```python
plt.xlabel("Days")
plt.ylabel("Temprature")
plt.title("Weather Chart")
plt.plot(x, y, color="Green", linestyle="dashed", marker="+")
```

```python
company = ["Google", "Amazon", "Freshworks", "Quadratyx"]
revenue = [90, 136, 89, 87]
profit = [40, 2, 34, 12]
```

```python
company_x_pos = np.arange(len(company))
```

```python
plt.xlabel("Company")
plt.ylabel("Revenue")
plt.title("US Based Company Stocks")
plt.xticks(company_x_pos, company)
plt.bar(company_x_pos-0.2, revenue, width=0.4,label="Company")
plt.bar(company_x_pos+0.2, profit, width=0.4,label="Profit")
plt.legend()
# It overlaps due to x pos of company
```
